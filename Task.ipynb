```python
#First import the specific variables needed for the task
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)

```

```python
#Importing the excel files into the database
df1= pd.read_excel(r"C:\Users\Leon's Laptop\Documents\Cowen\Client_Margin_Report_WithClass_TaskDay1.xlsx")
df2= pd.read_excel(r"C:\Users\Leon's Laptop\Documents\Cowen\Client_Margin_Report_WithClass_TaskDay2.xlsx")
```

```python
#Formatting the columns to the specific types in order to make the data clearer. As well as this, changing to 4 decimal places for accuracy
df1['Quantity']= df1['Quantity'].astype('int64')
df1['Value \n(Local)']= df1['Value \n(Local)'].astype('int64')
df1['Exposure \n(Local)']= df1['Exposure \n(Local)'].astype('int64')
df1['Exposure \n(USD)']= df1['Exposure \n(USD)'].astype('int64')
df1['Exposure \n(Local)']= df1['Exposure \n(Local)'].astype('int64')

df2['Quantity']= df1['Quantity'].astype('int64')
df2['Value \n(Local)']= df1['Value \n(Local)'].astype('int64')
df2['Exposure \n(Local)']= df1['Exposure \n(Local)'].astype('int64')
df2['Exposure \n(USD)']= df1['Exposure \n(USD)'].astype('int64')
df2['Exposure \n(Local)']= df1['Exposure \n(Local)'].astype('int64')
    
pd.set_option('display.float_format', '{:.4f}'.format)
```

```python
df1.head()
```

```python
df2.head()
```

```python
#In order to create a side by side comparison, we merge both the tables together using the common key 'Instrument'
merged_df= pd.merge(df1, df2, on='Instrument')
merged_df.head()
```

```python
#In order to show the changes in the specified columns, we use subtract dataframe 2 from dataframe 1 in order to show the changes. We also create new columns with the new headers
merged_df['Quantity change'] = merged_df['Quantity_x'] - merged_df['Quantity_y']
merged_df['Price Change'] = merged_df['Unit Price\n(Local)\n_x'] - merged_df['Unit Price\n(Local)\n_y']
merged_df['Local Value Change'] = merged_df['Value \n(Local)_x'] - merged_df['Value \n(Local)_y']
merged_df['Local Exposure Change'] = merged_df['Exposure \n(Local)_x'] - merged_df['Exposure \n(Local)_y']
merged_df['USD Value Change'] = merged_df['FX Rate_x'] - merged_df['FX Rate_y']
merged_df['Requirement Change'] = merged_df['Requirement \n(USD)_x'] - merged_df['Requirement \n(USD)_y']
merged_df['%Req Change'] = merged_df['%Req_x'] - merged_df['%Req_y']

merged_df
```

```python
#Dropping columns that aren't relevant
merged_df = merged_df.drop('Code_y', axis=1)
merged_df = merged_df.drop('Position Type_y', axis=1)
merged_df = merged_df.drop('Price Ccy_y', axis=1)
merged_df = merged_df.drop('Instrument Class_y', axis=1)
merged_df.head()
```

```python
#Renaming the columns to make it easier to understand
merged_df.rename(columns={'Code_x': 'Code', 'Position Type_x': 'Position Type', 'Quantity_x': 'Quantity (Day 1)', 'Unit Price\n(Local)\n_x': 'Unit Price in GDP (Day 1)','Value \n(Local)_x': 'Value in GDP (Day 1)', 'Exposure \n(Local)_x': 'Exposure in GDP (Day 1)','Price Ccy_x': 'Price Ccy', 'FX Rate_x': 'FX Rate (Day 1)','Instrument Class_x': 'Instrument Class', 'Exposure \n(USD)_x': 'Exposure in USD (Day 1)','Requirement \n(USD)_x': 'Requirement in USD (Day 1)', '%Req_x': '%Req (Day 1)'}, inplace=True)
merged_df.rename(columns={'Quantity_y': 'Quantity (Day 2)', 'Unit Price\n(Local)\n_y': 'Unit Price in GDP (Day 2)','Value \n(Local)_y': 'Value in GDP (Day 2)', 'Exposure \n(Local)_y': 'Exposure in GDP (Day 2)','FX Rate_y': 'FX Rate (Day 2)','Exposure \n(USD)_y': 'Exposure in USD (Day 2)','Requirement \n(USD)_y': 'Requirement in USD (Day 2)', '%Req_y': '%Req (Day 2)'}, inplace=True)
merged_df.head()
```

```python
merged_df = merged_df.loc[:,['Instrument','Instrument Class', 'Code', 'Price Ccy', 'Position Type', 'Quantity (Day 1)', 'Quantity (Day 2)', 'Quantity change', 'Unit Price in GDP (Day 1)', 'Unit Price in GDP (Day 2)', 'Price Change', 'Value in GDP (Day 1)', 'Value in GDP (Day 2)', 'Local Value Change', 'Exposure in GDP (Day 1)', 'Exposure in GDP (Day 2)', 'Local Exposure Change', 'Exposure in USD (Day 1)', 'Exposure in USD (Day 2)', 'USD Value Change', 'Requirement in USD (Day 1)','Requirement in USD (Day 2)','Requirement Change', '%Req (Day 1)','%Req (Day 2)', '%Req Change']]
```
